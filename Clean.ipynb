```python
import pandas as pd
import numpy as np

def clean_data(df):
	feature_columns = ["date_GMT","attendance", "home_team_name", "away_team_name", "Pre-Match PPG (Home)", "Pre-Match PPG (Away)"
	                ,"home_ppg", "away_ppg","home_team_goal_count", "away_team_goal_count", "home_team_goal_count_half_time", "away_team_goal_count_half_time"
	                ,"home_team_corner_count","away_team_corner_count","home_team_yellow_cards", "away_team_yellow_cards"
	                ,"home_team_red_cards", "away_team_red_cards", "home_team_shots", "away_team_shots", "home_team_shots_on_target", "away_team_shots_on_target"
	                ,"home_team_fouls", "away_team_fouls", "home_team_possession","away_team_possession"]
	n_features = len(feature_columns) # should equal to 26 when we add the result feature vector
	# print(n_features)
	cleaned_data = df[feature_columns]
	return cleaned_data

filename = "england-premier-league-matches-2018-to-2019-stats"
df = pd.read_csv(filename + ".csv");

df = clean_data(df);

# print(df.info())
```

```python


# name = "Manchester United"
# team_data[name]
```

```python
team_names = df.home_team_name.unique()

team_data = {}

for team in team_names:
	team_data[team] = { 
		"games_played" : 0 ,
		"goals_scored" : 0,
		"total_shots" : 0,
		"total_shots_on_target" : 0,
		"total_shots_stopped" : 0,
		"goals_conceded": 0,  
		"total_possession": 0,
		"total_fouls": 0,
		"total_yellow_cards": 0,
		"total_red_cards": 0,
		"total_wins": 0,
		"total_draws": 0,
		"total_losses": 0,
		"total_points": 0 
	}


home_columns = ["home_games_played", "home_goals_scored", "home_total_shots", "home_total_shots_on_target",
			   "home_total_shots_stopped", "home_goals_conceded", "home_total_possession", "home_total_fouls", "home_total_yellow_cards", "home_total_red_cards",
			   "home_total_wins", "home_total_draws", "home_total_losses", "home_total_draws"]

away_columns = ["away_games_played", "away_goals_scored", "away_total_shots", "away_total_shots_on_target",
			   "away_total_shots_stopped", "away_goals_conceded", "away_total_possession", "away_total_fouls", "away_total_yellow_cards", "away_total_red_cards",
			   "away_total_wins", "away_total_draws", "away_total_losses", "away_total_draws"]

def update_team_stats(df_row):
	home = df_row["home_team_name"]
	away = df_row["away_team_name"]
	home_goals = df_row["home_team_goal_count"]
	away_goals = df_row["away_team_goal_count"]
	
	team_data[home]["games_played"] += 1 
	team_data[home]["goals_scored"] += home_goals
	team_data[home]["total_shots"] += df_row["home_team_shots"] 
	team_data[home]["total_shots_on_target"] += df_row["home_team_shots_on_target"] 
	team_data[home]["total_shots_stopped"] += df_row["away_team_shots_on_target"] - df_row["away_team_goal_count"] 
	team_data[home]["goals_conceded"] += df_row["away_team_goal_count"] 
	team_data[home]["total_possession"] += df_row["home_team_possession"]
	team_data[home]["total_fouls"] += df_row["home_team_fouls"]
	team_data[home]["total_yellow_cards"] += df_row["home_team_yellow_cards"]
	team_data[home]["total_red_cards"] += df_row["home_team_red_cards"]
	
	team_data[away]["games_played"] += 1 
	team_data[away]["goals_scored"] += away_goals
	team_data[away]["total_shots"] += df_row["away_team_shots"] 
	team_data[away]["total_shots_on_target"] += df_row["away_team_shots_on_target"] 
	team_data[away]["total_shots_stopped"] += df_row["home_team_shots_on_target"] - df_row["home_team_goal_count"] 
	team_data[away]["goals_conceded"] += df_row["home_team_goal_count"] 
	team_data[away]["total_possession"] += df_row["away_team_possession"]
	team_data[away]["total_fouls"] += df_row["away_team_fouls"]
	team_data[away]["total_yellow_cards"] += df_row["away_team_yellow_cards"]
	team_data[away]["total_red_cards"] += df_row["away_team_red_cards"]

	if home_goals > away_goals:
		team_data[home]["total_wins"] += 1
		team_data[away]["total_losses"] += 1
		team_data[home]["total_points"] += 3
	elif home_goals == away_goals:
		# if the match was a draw
		# team_data[home]["total_draws"] += 1
		team_data[away]["total_draws"] += 1
	else:
		team_data[away]["total_wins"] += 1
		team_data[away]["total_points"] += 3
		team_data[home]["total_losses"] += 1


	print(f"{home} {home_goals} - {away_goals} {away}") 
	# team_data[home]["total_red_cards"] += df_row["total_red_cards"]
	# team_data[home]["shots_saved"] += df_row["away_team_shots"] - df_row["away"] 




```

```python
# sort data according to date
df.sort_values("date_GMT", inplace=True)
# checking that the dates are in fact sorted
for i in range(len(df)):
    print(df.loc[i, "date_GMT"])

```

Add empty columns to the dataframe

```python
# insert column into data frame
num_rows = len(df)
empty_col = np.zeros(num_rows, dtype="int")
new_columns = home_columns[:]

print(new_columns)

for cols in new_columns:
    df[cols] = empty_col
```

Now to add the lagging indicator features to the dataframe

```python
# now we're going to add lagging indicators to the dataset

for i in range(len(df)):
    # get the current fixture
    home, away = df.loc[i,["home_team_name", "away_team_name"]]
    # insert current current into running dataframe
    for col in home_columns:
        colname = col.replace("home_", "")
        df.loc[i, col] = team_data[home][colname]

    for col in away_columns:
        colname = col.replace("away_", "")
        df.loc[i, col] = team_data[away][colname]
        # df.loc[i,col]= team_data[team][col]


    # update current statics
    update_team_stats(df.loc[i])
```

```python
df['Result'] = df.apply(lambda row: 'H' if row['home_team_goal_count'] > row['away_team_goal_count']
                        else 'A' if row['home_team_goal_count'] < row['away_team_goal_count']
                        else 'D', axis=1)
```

```python
df.to_csv(filename + "(cleaned).csv")
```
